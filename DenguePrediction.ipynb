```python
#Read the dataset
import pandas as pd
import matplotlib.pyplot as plt
dengue_train_features = pd.read_csv("K:\\Dhwani\\1stSem\\ML\\Final\\Data\\Data\\DengAI_Predicting_Disease_Spread_-_Training_Data_Features.csv")
dengue_train_labels = pd.read_csv("K:\\Dhwani\\1stSem\\ML\\Final\\Data\\Data\\DengAI_Predicting_Disease_Spread_-_Training_Data_Labels.csv")


result=pd.concat([dengue_train_features, dengue_train_labels], axis=1, join='inner')

ProcessedData = result.apply(lambda x: x.fillna(method='ffill'))
ProcessedData.pop('week_start_date')

# Kelvin to celsius 
c=["reanalysis_air_temp_k", "reanalysis_avg_temp_k","reanalysis_dew_point_temp_k",
            "reanalysis_max_air_temp_k","reanalysis_min_air_temp_k"]
for i in c:
    ProcessedData[i] = ProcessedData[i] - 273.15
    
corr=ProcessedData.corr()

import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(10, 10))
plt.imshow(corr, cmap='RdYlGn', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns,rotation='vertical');
plt.yticks(range(len(corr.columns)), corr.columns);
plt.show()

result.hist(bins=20)
plt.show()
```
